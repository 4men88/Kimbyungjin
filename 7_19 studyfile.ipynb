### regular expression 정규식

```python
import re
```

```python
content = "Hello World"
dot = re.compile("(H..).(o..)")
plus = re.compile("l+")
```

```python
re.search(dot, content)
```

```python
re.search(plus, content)
```

```python
print(re.findall(plus, content))
```

```python
data = """
park = 800905-1049118
kim 700905-1059119
"""

pattern = re.compile("(\d{6})[-](\d{7})")

print(pattern, data)
print(re.findall(pattern, data))
print(pattern.sub("\g<1>-*******", data))
```

```python
p = re.compile("Crow|servo")
```

```python
m = p.match("CrowHello")
```

```python
print(m)
```

```python
re.search('^Life', 'Life is too short')
```

```python
re.search('Life$', 'My Life')
```

```python
p = re.compile("(ABC)+")
```

```python
m = p.search('ABCABCABC OK?')
print(m)
```

```python
p = re.compile(r"\bclass\b")
```

```python
re.search(p, "no class at all")
```

```python
p.search("no subclass at all")
```

```python
result = re.search(pattern, data)
```

```python
result.group(1)
```

```python
p = re.compile(r"(\w+)\s+\d+[-]\d+[-]\d+")
```

```python
m = re.search(p, "park 010-0000-0000")
```

```python
p = re.compile(r"(\w+) (\w+)")
result = re.search(p, "Issac Newton, physicist")
print(result.group())
print(result.group(1), result.group(2), result.group(1,2))
```

```python
p.sub("\g<2> \g<1>", "Isaac Newton, physicist")
```

```python
data = """
acccb
a....b
aaab
a.cccb
"""
p = re.compile("a[.]{3,}b")

p = re.compile("[.]")
```

```python
re.search(p,data).group()
```

```python
data = """
park 010-9999-9988
kim 010-9909-7789
lee 010-8789-7768
"""
p = re.compile(r"(\w+)(\s+)(\d+)[-](\d+)[-](\d+)")
result = p.findall(data)
print(p.sub("\g<1> \g<3>-\g<4>-####",data))
```

```python
data = """
park@naver.com
kim@daum.net
lee@myhome.co.kr
"""

p = re.compile(".*[@].*[.][^(com)|(net)]")
p.findall(data)
```

```python
p = re.compile(r"\W")
```
