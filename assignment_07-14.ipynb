```python
#61  Exercise 61:Average
a=[]
input_num=0
sum=0


while True:
    input_num=int(input())
       
    if input_num is 0:
        break
    a.append(input_num)
    
for i in a:
    sum += i

print(sum/len(a))
```

```python
#80
import random

k=0
avg=0
for k in range(0,10):
    num=5
    h=0   #h=1
    t=0   #t=0
    cnt=[]
    str=[]
    while True:
        num=random.randrange(0,2)
        cnt.append(num)
        if num:
            h+=1
            t=0
        else:
            t+=1
            h=0

        if h==3 or t==3:
            break

    for j in cnt:
        if cnt[j]:
            str.append('H')
        else:
            str.append('T')
    
    avg+=len(str)
    print(str,'(',len(str),'flips',')')
k+=1

print('On average, ',avg/10, 'flips were needed')
```
